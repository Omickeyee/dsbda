```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Iris.csv',header=None)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
```

```python
df.head()
```

# Create data subset for different species

```python
df['Species'].unique()
```

```python
df_setosa=df[df['Species']=='Iris-setosa']
df_setosa.head()
```

```python
df_versicolor=df[df['Species']=='Iris-versicolor']
df_versicolor.head()
```

# Merge two subsets

```python
df_merge=pd.concat([df_setosa,df_versicolor])
```

```python
df_merge
```

# Sort Data Petal Length

```python
df_sort=df.sort_values('PetalLengthCm',ascending = False)
df_sort.head()
```

# Transposing Data

```python
df_transpose=df.T
df_transpose
```

# Melting Data to long format

```python
df_melt=pd.melt(df,id_vars=['SepalLengthCm','SepalWidthCm','Species'])
df_melt.head()
```

# Casting data to wide format

```python
df_cast=pd.pivot_table(df,index=['Species'])
df_cast.head()
```
