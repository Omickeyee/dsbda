```python
soup= bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews = soup.findAll('div',{'class': 't-ZTKy'})
all_review=[]
for review in reviews:
    reviewText=review.get_text().replace('READ MORE',"").strip()
    all_review.append(reviewText)
    print(reviewText+"\n\n")
```

```python
ratings = soup.findAll('div',{'class':'_3LWZlK _1BLPMq'})
all_ratings=[]
for rating in ratings:
    all_ratings.append(rating.get_text())
all_ratings
```

```python
len(all_ratings)
```

```python
tags=soup.findAll('p',{'class':'_2-N8zT'})
all_tags=[]
for tag in tags:
    all_tags.append(tag.get_text())
all_tags
```

```python
len(all_tags)
```

```python
cus_names=soup.findAll('p',{'class':'_2sc7ZR _2V5EHH'})
all_cus_names=[]
for cus_name in cus_names:
    all_cus_names.append(cus_name.get_text())
all_cus_names
```

```python
data={
    "names":all_cus_names,
    "ratings":all_ratings,
    "tags":all_tags,
   
    
}
data
```

```python
pd.set_option('display.max_colwidth',None)
df=pd.DataFrame(data)
df
```
