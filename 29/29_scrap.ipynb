```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
link='https://www.amazon.in/Echo-Dot-3rd-Gen/product-reviews/B07PFFMP9P/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'
}
page = requests.get(link, headers=headers)
```

```python
soup = bs(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span', class_='a-profile-name')
```

```python
names
```

```python
cust_name = []

for i in range(len(names)):
  cust_name.append(names[i].get_text())
cust_name
```

```python
title = soup.find_all('a', class_='review-title-content')
```

```python
title
```

```python
review_title = []
for i in range(0,len(title)):
  review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
len(review_title)
```

```python
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate = []
for i in range(0,len(rating)):
 rate.append(rating[i].get_text())
rate
```

```python
len(rate)
```

```python
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
review_content = []
for i in range(0,len(review)):
 review_content.append(review[i].get_text())
review_content
```

```python
review_content[:] = [reviews.strip('\n') for reviews in review_content]
review_content
```

```python
len(review_content)
```

```python
df = pd.DataFrame()
df['Customer Name']=cust_name[:10]
df['Review title']=review_title[:10]
df['Ratings']=rate[:10]
df['Reviews']=review_content[:10] 
```

```python
df.head()
```

```python
df.to_csv(r'./reviews.csv',index=True)
```
